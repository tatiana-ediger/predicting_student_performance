```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# Reads data
students = pd.read_csv("student/student-all.csv")
```

```python
students_df = pd.DataFrame(students)
students_df
```

```python
apart_port_avg = students_df.loc[(students_df["Pstatus"] == "A") & (students_df["Math"] == 0)]["G3"].mean()
apart_math_avg = students_df.loc[(students_df["Pstatus"] == "A") & (students_df["Math"] == 1)]["G3"].mean()
together_port_avg = students_df.loc[(students_df["Pstatus"] == "T") & (students_df["Math"] == 0)]["G3"].mean()
together_math_avg = students_df.loc[(students_df["Pstatus"] == "T") & (students_df["Math"] == 1)]["G3"].mean()
```

```python
fig, ax = plt.subplots()
ind = np.array([1,2])
width = 0.35
plt1 = ax.bar(ind, [apart_port_avg, together_port_avg], width)
plt2 = ax.bar(ind + width, [apart_math_avg, together_math_avg], width)
ax.set_title("Class Average Grade by Parental Status")
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(("Portugese", "Math"))
ax.legend((plt1[0],plt2[0]), ("Apart", "Together"))
plt.ylim(top = 15)
plt.show()

# As we can see in this plot, there appears to be very little difference in final grade of students whose parents live together vs. apart
```
