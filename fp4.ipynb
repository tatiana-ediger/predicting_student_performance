```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# Reads data
students = pd.read_csv("student/student-all.csv")
```

```python
students_df = pd.DataFrame(students)
students_df
```

```python
students_df.columns
```

```python
# Calculates average grade by class
math_avg = students_df.loc[students_df["Math"] == 1]["G3"].mean()
port_avg = students_df.loc[students_df["Math"] == 0]["G3"].mean()
```

```python
# Plots bar graph for classes
plt.bar(["Portugese","Math"], [port_avg, math_avg])
plt.title("Grades by Class")
plt.show()
```

```python
# Calculates average grade by class, gender
female_port_avg = students_df.loc[(students_df["sex"] == "F") & (students_df["Math"] == 0)]["G3"].mean()
female_math_avg = students_df.loc[(students_df["sex"] == "F") & (students_df["Math"] == 1)]["G3"].mean()
male_port_avg = students_df.loc[(students_df["sex"] == "M") & (students_df["Math"] == 0)]["G3"].mean()
male_math_avg = students_df.loc[(students_df["sex"] == "M") & (students_df["Math"] == 1)]["G3"].mean()
```

```python
# Plots grouped bar chart for class and gender
fig, ax = plt.subplots()
ind = np.array([1,2])
width = 0.35
plt1 = ax.bar(ind, [female_port_avg, male_port_avg], width)
plt2 = ax.bar(ind + width, [female_math_avg, male_math_avg], width)
ax.set_title("Class Average Grade by Gender")
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(("Portugese", "Math"))
ax.legend((plt1[0],plt2[0]), ("Female", "Male"))
plt.ylim(top = 15)
plt.show()
```

```python
# Calculates average grades by school and class
GP_port_avg = students_df.loc[(students_df["school"] == "GP") & (students_df["Math"] == 0)]["G3"].mean()
GP_math_avg = students_df.loc[(students_df["school"] == "GP") & (students_df["Math"] == 1)]["G3"].mean()
MS_port_avg = students_df.loc[(students_df["school"] == "MS") & (students_df["Math"] == 0)]["G3"].mean()
MS_math_avg = students_df.loc[(students_df["school"] == "MS") & (students_df["Math"] == 1)]["G3"].mean()
```

```python
# Plots grouped bara chart for class, school
fig, ax = plt.subplots()
ind = np.array([1,2])
width = 0.35
plt1 = ax.bar(ind, [GP_port_avg, MS_port_avg], width)
plt2 = ax.bar(ind + width, [GP_math_avg, MS_math_avg], width)
ax.set_title("Class Average Grade by School")
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(("Portugese", "Math"))
ax.legend((plt1[0],plt2[0]), ("Gabriel Perreira", "Mousinho da Silva"))
plt.ylim(top = 15)
plt.show()
```
