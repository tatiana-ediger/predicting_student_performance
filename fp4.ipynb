```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Reads data
students = pd.read_csv("student/student-all.csv")
```

```python
students_df = pd.DataFrame(students)
students_df
```

```python
a_status_avg = students_df.loc[students_df["Pstatus"] == "A"]["G3"].mean()
t_status_avg = students_df.loc[students_df["Pstatus"] == "T"]["G3"].mean()
```

```python
plt.bar(["A","T"], [a_status_avg,t_status_avg])

# As we can see in this plot, there appears to be very little difference in final grade of students whose parents live together vs. apart
```

```python
students_df.plot(kind="scatter", x="absences", y="G3")
```

```python
math_avg = students_df.loc[students_df["Math"] == 1]["G3"].mean()
port_avg = students_df.loc[students_df["Math"] == 0]["G3"].mean()
```

```python
plt.bar(["Math","Portugese"], [math_avg, port_avg])
```

```python
int_avg = students_df.loc[students_df["internet"] == "yes"]["G3"].mean()
no_int_avg = students_df.loc[students_df["internet"] == "no"]["G3"].mean()
```

```python
plt.bar(["Internet", "No Internet"], [int_avg, no_int_avg])
```

```python
female_avg = students_df.loc[students_df["sex"] == "F"]["G3"].mean()
male_avg = students_df.loc[students_df["sex"] == "M"]["G3"].mean()
```

```python
plt.bar(["Female", "Male"], [female_avg, male_avg])
```
